#
AA Average Frequency, = The average number of times households or persons viewed a given program, station, or advertisement during a specific time period, such as one month. This number is derived by dividing the Gross Rating Points (GRP) by the total non-duplicated audience (cume). For example, if a group of programs achieved 30 GRPs and a cume of 20, then the average frequency would be 1.5 exposures per person or household.

##find the genre that frequently occurs with high IMDB ratings for each service, you can count the occurrences of each genre for each service and then select the genre with the highest count of high IMDB ratings.

```python
# Filter for specific services
services_of_interest = ['netflix', 'hbo_max', 'hbo', 'hulu_plus', 'peacock', 'amazon_prime', 'disney_plus', 'fox']
reelgood_data_filtered = reelgood_data[reelgood_data['Service'].isin(services_of_interest)]

# Drop rows with NaN values in 'IMDB' column
reelgood_data_filtered.dropna(subset=['IMDB'], inplace=True)

# Define a threshold for high IMDB ratings
imdb_threshold = 8.0

# Filter for rows with IMDB ratings above the threshold
high_imdb_data = reelgood_data_filtered[reelgood_data_filtered['IMDB'] > imdb_threshold]

# Group by 'Service' and 'Genre', and count occurrences of high IMDB ratings for each genre
genre_imdb_counts = high_imdb_data.groupby(['Service', 'Genre']).size().reset_index(name='Count')

# Find the genre with the highest count of high IMDB ratings for each service
max_count_indices = genre_imdb_counts.groupby('Service')['Count'].idxmax()
top_genre_per_service = genre_imdb_counts.loc[max_count_indices]

print(top_genre_per_service)
```

## find the genre that occurs most frequently with low IMDB ratings for each service, you can count the occurrences of each genre for each service and then select the genre with the highest count of low IMDB ratings.

```python
# Filter for specific services
services_of_interest = ['netflix', 'hbo_max', 'hbo', 'hulu_plus', 'peacock', 'amazon_prime', 'disney_plus', 'fox']
reelgood_data_filtered = reelgood_data[reelgood_data['Service'].isin(services_of_interest)]

# Drop rows with NaN values in 'IMDB' column
reelgood_data_filtered.dropna(subset=['IMDB'], inplace=True)

# Define a threshold for low IMDB ratings
imdb_threshold = 5.0

# Filter for rows with IMDB ratings below the threshold
low_imdb_data = reelgood_data_filtered[reelgood_data_filtered['IMDB'] < imdb_threshold]

# Group by 'Service' and 'Genre', and count occurrences of low IMDB ratings for each genre
genre_imdb_counts = low_imdb_data.groupby(['Service', 'Genre']).size().reset_index(name='Count')

# Find the genre with the highest count of low IMDB ratings for each service
max_count_indices = genre_imdb_counts.groupby('Service')['Count'].idxmax()
top_genre_per_service = genre_imdb_counts.loc[max_count_indices]

print(top_genre_per_service)
```

```python
def recommend_based_on_input(input_text, reelgood_data):
    # Convert input text to lowercase for case-insensitive matching
    input_text = input_text.lower()

    # Filter the DataFrame based on the input text
    filtered_data = reelgood_data[reelgood_data['Title'].str.lower().str.contains(input_text) |
                         reelgood_data['Genre'].str.lower().str.contains(input_text)]


    filtered_data = filtered_data.sort_values(by='IMDB', ascending=False)


    recommendations = filtered_data.head(5)[['Title', 'IMDB', 'ReelGood']]

    return recommendations


user_input = input("Please provide a genre of movies or shows you like: ")

recommendations = recommend_based_on_input(user_input, reelgood_data)

if recommendations.empty:
    print("Sorry, no recommendations found based on your input.")
else:
    print("Recommendations based on your input:")
    print(recommendations)
```

#Rank the recommendations by both IMDB rating and ReelGood rating,

```python
def recommend_based_on_input(input_text, reelgood_data):
    # Convert input text to lowercase for case-insensitive matching
    input_text = input_text.lower()


    filtered_data = reelgood_data[reelgood_data['Title'].str.lower().str.contains(input_text) |
                                  reelgood_data['Genre'].str.lower().str.contains(input_text)]


    filtered_data = filtered_data.sort_values(by=['IMDB', 'ReelGood'], ascending=[False, False])

    filtered_data = filtered_data.drop_duplicates(subset='Title')


    filtered_data = filtered_data.dropna(subset=["What it's about", 'Where to Watch'])


    recommendations = filtered_data.head(5)[['Title', 'IMDB', 'ReelGood', "What it's about", 'Where to Watch']]

    return recommendations


user_input = input("Please provide a genre of movies or shows you like: ")

recommendations = recommend_based_on_input(user_input, reelgood_data)

if recommendations.empty:
    print("Sorry, no recommendations found based on your input.")
else:
    print("Recommendations based on your input:")
    print(recommendations)
```
